## 10. Feature Engineering

### Purpose

Feature engineering is the process of creating new variables from the existing dataset to improve the performance of machine learning models. These engineered features capture meaningful transaction behaviors that are not directly available in the raw dataset.

For financial fraud detection, derived features such as balance differences, transaction ratios, time-based attributes, and high-value transaction indicators help machine learning algorithms identify complex fraud patterns more effectively than relying solely on the original variables.

The features created in this section will be used in the predictive modeling phase to improve fraud detection accuracy and support business decision-making.

### 10.1 Create Time-Based Features

### Purpose

The PaySim dataset records transactions using the **step** variable, where each step represents one hour of simulated time. Converting this variable into meaningful time-based features allows us to analyze transaction behavior across different hours and days. These features may help identify temporal fraud patterns and improve model performance.

```python
# ============================================================
# Create Time-Based Features
# ============================================================

# Hour of the day (0–23)
sample_df["hour"] = sample_df["step"] % 24

# Day number
sample_df["day"] = (sample_df["step"] // 24) + 1

# Display sample records
sample_df[["step", "hour", "day"]].head()
```

### Observation

Two new time-based features, **hour** and **day**, were successfully derived from the **step** variable. The **hour** feature represents the hour of the day (0–23), while the **day** feature identifies the corresponding simulation day.

The sample output confirms that the conversion has been performed correctly. For example, transactions occurring at simulation step **400** are mapped to **Hour 16** on **Day 17**, while step **136** corresponds to **Hour 16** on **Day 6**. These engineered features preserve the chronological information contained in the original dataset while making it more interpretable for analysis and predictive modeling.

### Interpretation

Time-based transaction behavior often provides valuable information for fraud detection. Fraudulent transactions may exhibit patterns during specific hours or simulation periods that are not directly observable from the raw **step** variable. Transforming the simulation step into meaningful temporal attributes improves interpretability and enables the machine learning model to learn potential time-dependent fraud patterns.

### 10.2 Create Balance Difference Features

### Purpose

Balance difference features measure how account balances change before and after each transaction. These variables help capture transaction behavior by quantifying the amount deducted from the sender's account and the amount credited to the recipient's account. Significant balance changes may indicate unusual transaction patterns associated with fraudulent activities.

```python
# ============================================================
# Create Balance Difference Features
# ============================================================

sample_df["balanceOrigDiff"] = (
    sample_df["oldbalanceOrg"] -
    sample_df["newbalanceOrig"]
)

sample_df["balanceDestDiff"] = (
    sample_df["newbalanceDest"] -
    sample_df["oldbalanceDest"]
)

sample_df[
    [
        "oldbalanceOrg",
        "newbalanceOrig",
        "balanceOrigDiff",
        "oldbalanceDest",
        "newbalanceDest",
        "balanceDestDiff"
    ]
].head()
```

### Observation

Two balance difference features, **balanceOrigDiff** and **balanceDestDiff**, were successfully generated by calculating the changes in sender and recipient account balances before and after each transaction.

The sample output demonstrates different transaction scenarios. In several transactions, the sender's balance decreases while the recipient's balance increases by approximately the same amount, reflecting normal money transfer behavior. In other cases, negative balance differences are observed, indicating transactions where account balances increase rather than decrease. These variations capture valuable financial transaction patterns that are not explicitly represented in the original dataset.

### Interpretation

Balance difference features provide a quantitative representation of money movement during each transaction. Rather than relying solely on raw account balances, these engineered variables capture the actual financial impact of each transaction, allowing anomaly detection models to identify unusual balance changes that may indicate fraudulent activities.

### 10.3 Create Transaction Ratio Features

### Purpose

Transaction ratio features measure the relationship between the transaction amount and the sender's available account balance. Instead of considering only the absolute transaction amount, these features help determine whether a transaction represents a normal financial activity or an unusually large proportion of the customer's available funds. Such behavioral indicators can significantly improve fraud detection performance.

```python
# ============================================================
# Create Transaction Ratio Feature
# ============================================================

sample_df["amount_to_balance_ratio"] = (
    sample_df["amount"] /
    (sample_df["oldbalanceOrg"] + 1)
)

sample_df[
    [
        "amount",
        "oldbalanceOrg",
        "amount_to_balance_ratio"
    ]
].head()
```

### Observation

The **amount_to_balance_ratio** feature was successfully generated by comparing the transaction amount with the sender's original account balance. The sample output demonstrates considerable variation across transactions. Most transactions exhibit relatively small ratio values (less than 3), indicating that the transferred amount is proportionate to the available account balance.

However, certain transactions display exceptionally high ratio values. For example, one transaction records a ratio of approximately **118.95**, indicating that the transaction amount is significantly larger than the sender's available balance. Such extreme values may represent unusual transaction behavior and require further investigation during fraud detection.

### Interpretation

The transaction-to-balance ratio provides a behavioral perspective that is more informative than considering transaction amount alone. Transactions with unusually high ratio values may indicate abnormal financial activity, especially when combined with other factors such as transaction type, account balance changes, and transaction timing. This engineered feature is expected to improve the machine learning model's ability to distinguish suspicious transactions from normal financial operations.

### 10.4 Create High-Value Transaction Indicator

### Purpose

High-value transactions often receive additional scrutiny in financial institutions because they may involve greater financial risk. This feature classifies transactions as high-value or normal-value based on the 95th percentile of transaction amounts within the dataset. The resulting binary indicator enables machine learning models to distinguish unusually large transactions from routine financial activities.

```python
# ============================================================
# Create High-Value Transaction Indicator
# ============================================================

threshold = sample_df["amount"].quantile(0.95)

sample_df["high_value_transaction"] = (
    sample_df["amount"] >= threshold
).astype(int)

print(f"95th Percentile Threshold: {threshold:,.2f}")

sample_df[
    [
        "amount",
        "high_value_transaction"
    ]
].head()
```

### Observation

The **high_value_transaction** feature was successfully generated using the **95th percentile** of transaction amounts as the threshold. Transactions with an amount greater than or equal to **518,661.13** were classified as high-value transactions, while all remaining transactions were categorized as normal-value transactions.

The sample output confirms that transactions exceeding the threshold are correctly assigned a value of **1**, whereas transactions below the threshold receive a value of **0**. This binary classification simplifies the identification of unusually large financial transactions that may require additional fraud monitoring.

### Interpretation

High-value transactions often represent increased financial exposure and therefore warrant additional scrutiny. Although not all high-value transactions are fraudulent, combining this indicator with other behavioral features such as transaction type, balance differences, transaction timing, and customer activity enables machine learning models to identify suspicious transaction patterns more effectively.

#### 10.4.1 Verify High-Value Transaction Distribution

```python
# ============================================================
# Verify High-Value Transaction Distribution
# ============================================================

sample_df["high_value_transaction"].value_counts()
```

### Observation

The distribution of the **high_value_transaction** feature confirms that the threshold has been applied correctly. Out of **500,000** sampled transactions, **475,000 (95%)** were classified as normal-value transactions, while **25,000 (5%)** were identified as high-value transactions based on the 95th percentile threshold.

This distribution aligns with the intended feature engineering strategy, ensuring that only the most significant transaction amounts are flagged for additional analysis.

### Interpretation

Flagging only the top 5% of transaction amounts enables the fraud detection system to focus on transactions with potentially higher financial risk while preserving the majority of normal transactions. This binary feature serves as an additional behavioral indicator that can complement other engineered variables during machine learning model training.

### 10.5 Create Zero Balance Indicators

### Purpose

Zero balance indicators identify transactions where either the sender's or the recipient's account has a balance of zero before the transaction. Such scenarios may reveal unusual account behavior and have been observed in financial fraud datasets. These binary features provide additional behavioral information that may improve the effectiveness of fraud detection models.

```python
# ============================================================
# Create Zero Balance Indicators
# ============================================================

sample_df["zero_balance_sender"] = (
    sample_df["oldbalanceOrg"] == 0
).astype(int)

sample_df["zero_balance_receiver"] = (
    sample_df["oldbalanceDest"] == 0
).astype(int)

sample_df[
    [
        "oldbalanceOrg",
        "oldbalanceDest",
        "zero_balance_sender",
        "zero_balance_receiver"
    ]
].head()
```

```python
print("Zero Balance Sender")
print(sample_df["zero_balance_sender"].value_counts())

print("\nZero Balance Receiver")
print(sample_df["zero_balance_receiver"].value_counts())
```

### Observation

The **zero_balance_sender** and **zero_balance_receiver** features were successfully created by identifying transactions where the sender's or recipient's account balance was zero before the transaction.

The sample output confirms that the binary indicators were assigned correctly. In the sampled dataset of **500,000** transactions, **165,541** transactions involved senders with a zero initial balance, while **212,002** transactions involved recipients with a zero initial balance. The remaining transactions had non-zero account balances prior to the transaction.

### Interpretation

Zero account balances can provide valuable behavioral information during fraud detection. Transactions involving accounts with zero balances may represent newly created accounts, inactive accounts, or unusual transaction scenarios. Although these conditions are not inherently fraudulent, combining these indicators with transaction amount, balance differences, transaction type, and temporal features enables machine learning models to better recognize suspicious financial behavior.
